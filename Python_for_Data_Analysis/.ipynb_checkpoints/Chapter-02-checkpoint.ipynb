```python
# the datetime api is pretty baroque. It alllows a lot, but every object is either date, time or datetime.
# Remember these because you will need them:

import datetime as dt
now = dt.datetime.now()

print(now.date())
print(now.time())

# And you can do arithmetic on these objects

print(now + dt.timedelta(minutes=334))

```

The signature of timedelta looks like this: 

timedelta(days=0, seconds=0, microseconds=0, milliseconds=0, minutes=0, hours=0, weeks=0)

It is a great way to add/subtract from a day, and also the way in which all differences in datetime objects is specified.


```python
# See how much time elapsed between the previous cell and this one.

n2 = dt.datetime.now()

print(n2 - now)
```

```python
# And you can print these objects using the standard C spec for formatting of dates (also what the date program uses)
now.strftime('%d/%m/%Y %H:%M')
```

```python
# 'pass' is needed in control blocks where indentation is required for code but you don't have any code

if (3>2):
    # Do nothing here
    pass
else:
    # Do something else
    print ("How did we ever get here?")
    
# Learning about yield

def allprimes(x):
    """ Returns all primes between 1 and x"""
    def isPrime(y):
        for m in range(2,y):
            if (y % m == 0):
                return False
        return True
    for i in range(x):
        if (isPrime(i)):
            yield i


# Now use this object, by 'calculating' all primes between 1 and 3 million
for i in allprimes(3000000000):
    print (i)
    # We could do this forever, but let's stop after a few.
    if i > 3030:
        break

    
```
